```python
from shapely import wkt
import geopandas as gpd

df2["geometry"] = df2["geometry"].apply(wkt.loads)
```

```python
gdf = gpd.GeoDataFrame(
    df2,
    geometry="geometry",
    crs="EPSG:4326"   # important si coords GPS
)
```

```python
type(gdf)
#gdf.crs
```

```python
departements = gpd.read_file("../data/shp/departements-20170102.shp", encoding="utf-8")

print(departements.head())
print(departements.crs)
```

```python
gdf_dep = gpd.sjoin(
    gdf,
    departements[["nom", "code_insee", "geometry"]],
    how="left",
    predicate="within"
)
```

```python
gdf_dep.head()
```

```python
mapping = pd.read_csv("../data/processed/taxref_species_vernacular_mapping.csv")

gdf_dep = gdf_dep.merge(mapping, on="species", how="left")
```

```python
gdf_dep.head()
```

```python
ep_sig = gpd.read_file("../data/shp/sig_metrop.shp", encoding="utf-8")
# si accents cassés: encoding="latin1"
```

```python
ep_site = pd.read_csv("../data/raw/ep_site.csv", sep=";", encoding="utf-8")
# adapte sep="," si besoin
```

```python
liste_des = pd.read_csv("../data/raw/liste_designations.csv", sep=";", encoding="utf-8")
```

```python
liste_des.head()
```

```python
ep = ep_sig.merge(ep_site, on="cd_sig", how="left")
```

```python
ep = ep.merge(liste_des, on="id_designation", how="left")
```

```python
ep.columns
```

```python
print(ep.head().to_string())
```

```python
# Harmoniser CRS
ep = ep.to_crs(gdf_dep.crs)
```

```python
j = gpd.sjoin(
    gdf_dep[["geometry"]],             # on ne garde que la géométrie côté gauche
    ep[["cd_sig", "nom", "lb_designation","abr_designation_sig","categorie_uicn", "geometry"]],  # adapte "nom"/"designation"
    how="left",
    predicate="within"  # ou "intersects" si points sur frontières
)
```

```python
in_ep = j["index_right"].notna().groupby(level=0).any()
gdf_dep["in_ep"] = gdf_dep.index.to_series().map(in_ep).fillna(False)
```

```python
names = (
    j.dropna(subset=["nom"])
     .groupby(level=0)["nom"]
     .apply(lambda s: "; ".join(sorted(set(map(str, s)))))
)
gdf_dep["ep_names"] = gdf_dep.index.to_series().map(names)
```

```python
types_ = (
    j.dropna(subset=["lb_designation"])
     .groupby(level=0)["lb_designation"]
     .apply(lambda s: "; ".join(sorted(set(map(str, s)))))
)
gdf_dep["ep_types"] = gdf_dep.index.to_series().map(types_)
```

```python
types_abr = (
    j.dropna(subset=["abr_designation_sig"])
     .groupby(level=0)["abr_designation_sig"]
     .apply(lambda s: "; ".join(sorted(set(map(str, s)))))
)
gdf_dep["ep_types_abr"] = gdf_dep.index.to_series().map(types_abr)
```
